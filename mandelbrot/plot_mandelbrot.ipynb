```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
path = '/content/drive/MyDrive/UAM ACO' # path to the folder with the data file
```

```python
df = pd.read_csv(os.path.join(path, 'mandelbrot.dat'))
df = df[df.divergence==0] #selecting only not diverged points to plot
man = plt.scatter(x=df['c_real'], y=df['c_img'], marker='+', s=6)
plt.legend([man],
           ['Not diverged'],
           scatterpoints=1,
           loc='upper left',
           fontsize=12)
plt.title('Mandelbrot Set')
plt.xlim([-2,0.5]);
plt.ylim([-1.1,1.1]);
```
