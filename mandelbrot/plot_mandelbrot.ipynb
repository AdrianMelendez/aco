```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
path = '/content/drive/MyDrive/UAM ACO'
```

```python
df = pd.read_csv(os.path.join(path, 'mandelbrot.dat'))
df = df[df.divergence==0]
man = plt.scatter(x=df['c_real'], y=df['c_img'], marker='+')
plt.legend([man],
           ['Not diverged'],
           scatterpoints=1,
           loc='best',
           fontsize=8)
plt.xlim([-2,0.5])
```
